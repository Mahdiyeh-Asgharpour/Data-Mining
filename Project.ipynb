```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Project_Data.csv")
```

```python
df = df[:-2]
df.replace("NAN", np.nan, inplace=True)
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
for col in df.columns[1:]:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
print(df.head())
```

```python
df
```

```python
df.rename(columns={'Unnamed: 0': 'Name'}, inplace=True)
df
```

```python
numeric_columns = df.columns[1:]
means = df[numeric_columns].mean()
std_devs = df[numeric_columns].std()
z_scores = (df[numeric_columns] - means) / std_devs
outliers = (z_scores.abs() > 3)
print(f"Outliers:\n{outliers}")
```

```python
means = df.iloc[:, 1:].mean()
std_devs = df.iloc[:, 1:].std()
z_scores = (df.iloc[:, 1:] - means) / std_devs
outliers = (z_scores.abs() > 3)
print("Outliers identified by Z-score > 3:")
for col in outliers.columns:
    if outliers[col].any():
        print(f"\n{col} outliers:")
        print(df[outliers[col]][['Name', col]])
for col in df.columns[1:]:
    df.loc[outliers[col], col] = means[col]
print("\nDataFrame after handling outliers:")
df
```

```python
for col in numeric_columns:
    mean_value = df[col].mean()
    df.loc[outliers[col], col] = mean_value
print(f"Data after handling outliers:")
df
```

```python
numeric_columns = df.columns[1:]
means = df[numeric_columns].mean()
std_devs = df[numeric_columns].std()
for col in numeric_columns:
    df[col] = (df[col] - means[col]) / std_devs[col]
print("Data after standardization:")
df
```

```python
df.iloc[:,0]
```

```python
numeric_df = df.drop(columns=['Name'])
corr_matrix = numeric_df.corr().abs()
upper_triangle = np.triu(np.ones(corr_matrix.shape), k=1)
corr_values = corr_matrix.where(upper_triangle == 1)
features_removed = 0
while features_removed < 2:
    max_corr = corr_values.stack().idxmax()
    feature1, feature2 = max_corr
    label_corr = corr_matrix['Total'].abs()
    if label_corr[feature1] < label_corr[feature2]:
        numeric_df.drop(columns=[feature1], inplace=True)
    else:
        numeric_df.drop(columns=[feature2], inplace=True)
    corr_matrix = numeric_df.corr().abs()
    upper_triangle = np.triu(np.ones(corr_matrix.shape), k=1)
    corr_values = corr_matrix.where(upper_triangle == 1)
    features_removed += 1
    print("Removed Feature:", feature1 if label_corr[feature1] < label_corr[feature2] else feature2)
    print("Remaining Features:", numeric_df.columns)
df = pd.concat([df[['Name']], numeric_df], axis=1)
print("\nFinal DataFrame after removing two features:")
df
```
